```python
import pandas


data1 = pandas.read_csv("/content/New_final_balanced_sentiment_dataset.csv")
data2 = pandas.read_csv("/content/Team1_labelled_dataset.csv")
data3 = pandas.read_csv("/content/balanced_konkani_sentiment.csv")
data4 = pandas.read_excel("/content/final1ksentences.xlsx")
```

```python
data1.rename(columns = {'Sentiment':'label'}, inplace = True)
print(data1.columns)
```

```python
print(data2.columns)
```

```python
data3.rename(columns = {'Sentiment':'label'}, inplace = True)
data3.rename(columns = {'Konkani_sentences':'konkani'}, inplace = True)
print(data3.columns)
```

```python
data4.rename(columns = {'Sentiment':'label'}, inplace = True)
data4.rename(columns = {'Konkani Sentences':'konkani'}, inplace = True)
print(data4.columns)
```

```python
out_data_1 = data1[['konkani','label']]
out_data_2 = data2[['konkani','label']]
out_data_3 = data3[['konkani','label']]
out_data_4 = data4[['konkani','label']]
```

```python
# prompt: combine out_data_n into 1 dataframe and make all isntances of label as lower case and shuffle the rows

import pandas as pd

out_data = pd.concat([out_data_1, out_data_2, out_data_3, out_data_4], ignore_index=True)
out_data['label'] = out_data['label'].str.lower()
out_data = out_data.sample(frac=1).reset_index(drop=True)
out_data
```

```python
out_data.to_csv("/content/final_dataset_4000.csv")
```
