```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('final_dataset_4000.csv')
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("\nClass Sample Size:")
class_counts = df['label'].value_counts()
print(class_counts)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='label', data=df, palette='Set2')
plt.title('Class Distribution')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\nTotal Number of Classes:", df['label'].nunique())
```

```python
df['char_count'] = df['konkani'].apply(len)
df['word_count'] = df['konkani'].apply(lambda x: len(str(x).split()))
```

```python
print("\nDescriptive Stats (Character Count):")
print(df['char_count'].describe())
```

```python
print("\nDescriptive Stats (Word Count):")
print(df['word_count'].describe())
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['char_count'], kde=True, color='skyblue', bins=30)
plt.title('Character Count Distribution')
plt.xlabel('Number of Characters')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['word_count'], kde=True, color='salmon', bins=30)
plt.title('Word Count Distribution')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
